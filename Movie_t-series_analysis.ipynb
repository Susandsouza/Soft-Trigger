```python
# a pie chart
type_counts = df['type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#66b3ff'])
plt.title('Distribution of Movies and TV Shows')
plt.axis('equal') 
plt.show()
```

```python
print(df.columns)
unnamed_columns = [col for col in df.columns if 'Unnamed' in col]
print("Columns to drop:", unnamed_columns)

df.drop(columns=unnamed_columns, inplace=True)
print(df.columns)
df.hear
```

```python
#Handle Missing Data
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.groupby('country').agg(count=('country','count')).reset_index().sort_values(by='count',ascending=False)
```

```python
# Visualizing missing data as a heatmap
import seaborn as sns
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
# Histograms for all numeric columns
import matplotlib.pyplot as plt

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

if len(numeric_columns) > 0:
    for column in numeric_columns:
        plt.figure(figsize=(8, 5))
        plt.hist(df[column].dropna(), bins=10, edgecolor='black', color='skyblue')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
else:
    print("No numeric columns found for histogram visualization.")
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns
if len(numeric_columns) > 1:
    correlation_matrix = df[numeric_columns].corr()
    print("\nCorrelation Matrix:")
    print(correlation_matrix)
    
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix Heatmap')
    plt.show()
else:
    print("Not enough numeric columns for correlation analysis.")
```
